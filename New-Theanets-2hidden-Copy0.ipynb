```python
import math, string, random, itertools, copy
import numpy as np
from numpy import float32
import matplotlib.pyplot as plt
from scipy.stats import poisson
from copy        import deepcopy
```

```python
import climate
import theanets
import time
from sklearn.datasets import make_classification
from sklearn.metrics  import confusion_matrix
from theanets.trainer import SupervisedPretrainer
```

```python
maxJets  = 20
nInputs  = maxJets * 4 + 2                    # Maximum number of Jets x 4-momentum + METx,y              
nOutputs = 1                                  # Trigger Bit: 0 or 1  
                                              # 2 hidden layers
    
exp=theanets.Experiment(theanets.feedforward.Regressor(layers=(nInputs,nInputs, nInputs,(nOutputs, 'sigmoid'))),'layerwise')
```

```python
nEvents=420000                                     # Number of Events = 120000
nJetsPerEvent=np.random.poisson(6,size=nEvents)    # Number of Jets per event = Poisson(6)


target_matrix=np.zeros((nEvents,1))
input_matrix =np.zeros((nEvents, nInputs))

n_Mass = np.ndarray((nEvents, maxJets))
n_Phi  = np.ndarray((nEvents, maxJets))
n_Eta  = np.ndarray((nEvents, maxJets))
n_E    = np.ndarray((nEvents, maxJets))
n_Etx  = np.ndarray((nEvents, maxJets))
n_Ety  = np.ndarray((nEvents, maxJets))
n_E    = np.ndarray((nEvents, maxJets))
n_Et   = np.ndarray((nEvents, maxJets))
n_Px   = np.ndarray((nEvents, maxJets))
n_Py   = np.ndarray((nEvents, maxJets))
n_Pz   = np.ndarray((nEvents, maxJets))
n_Pt   = np.ndarray((nEvents, maxJets))
n_P    = np.ndarray((nEvents, maxJets))
n_Ex_miss=np.ndarray((nEvents))
n_Ey_miss=np.ndarray((nEvents))
n_Et_miss=np.ndarray((nEvents))


minMJ1=None
minMJ2=None

Px_V,  Py_V,  Pz_V,  Pt_V          = [],[],[],[]
Etmiss_V, Exmiss_V, Eymiss_V       = [],[],[]
Eta_Values  =[]
Phi_Values  =[]
Mass_Values =[]
E_Values    =[]
P_Values    =[]

R2_Values   =[]
R2_fit      =[]
MR_Values   =[]

nPass1=0
nFail1=0




for iEvent in range(nEvents):          #Loop over all events
    
    nJets=nJetsPerEvent[iEvent]  
    
    if(nJets <= 20): 
        Jets_P4=[]    
    
        n_Ex_miss[iEvent]=random.gauss(0,3)
        n_Ey_miss[iEvent]=random.gauss(0,3)
        n_Et_miss[iEvent]=sqrt(n_Ex_miss[iEvent]**2 + n_Ey_miss[iEvent]**2)
        MET= [n_Ex_miss[iEvent],n_Ey_miss[iEvent]]
    
        Etmiss_V.append(n_Et_miss[iEvent])
        Exmiss_V.append(n_Ex_miss[iEvent])
        Eymiss_V.append(n_Ey_miss[iEvent])    
    
    
        for iJet in range(nJets):   
        
            # Synthetic Data
            n_Mass[iEvent][iJet]  = random.expovariate(0.165) 
            n_Phi [iEvent][iJet]  = -math.pi + math.pi *2*random.random()
            n_Eta [iEvent][iJet]  = -3.0 + 2*3.0*random.random()
            n_Pt  [iEvent][iJet]  = 100*random.expovariate(2.0)        
            n_Px  [iEvent][iJet]  = random.gauss(0,3)        
            n_Py  [iEvent][iJet]  = random.gauss(0,3)
            n_Pz  [iEvent][iJet]  = random.gauss(0,6)     
        

            n_P   [iEvent][iJet]  = sqrt((n_Px[iEvent][iJet])**2 + (n_Py[iEvent][iJet])**2+(n_Pz[iEvent][iJet])**2)
        
            n_Etx [iEvent][iJet]  = random.gauss(0,3)
            n_Ety [iEvent][iJet]  = random.gauss(0,3) 
            n_Et  [iEvent][iJet]  = sqrt(n_Etx [iEvent][iJet]**2+ n_Ety [iEvent][iJet]**2)              
            n_E   [iEvent][iJet]  = sqrt((n_P[iEvent][iJet])**2  + (n_Mass[iEvent][iJet])**2)
       
            E_Values.append(n_E [iEvent][iJet])
            P_Values.append(n_P [iEvent][iJet])
            Px_V.append(n_Px  [iEvent][iJet])
            Py_V.append(n_Py  [iEvent][iJet])
            Pz_V.append(n_Pz  [iEvent][iJet])
            Pt_V.append(n_Pt  [iEvent][iJet])
        
            Eta_Values.append(n_Eta[iEvent][iJet])
            Phi_Values.append(n_Phi[iEvent][iJet])
            Mass_Values.append(n_Mass[iEvent][iJet])
        
            iJetP4=(n_E[iEvent][iJet],n_Px[iEvent][iJet],n_Py[iEvent][iJet],n_Pz[iEvent][iJet])
         
            Jets_P4.append(iJetP4)
        
        # ---------------------------------------------------------
        def getKey(item):
            return sqrt(item[1]*item[1]+item[2]*item[2])
        # ---------------------------------------------------------
    
        Jets_P4=sorted(Jets_P4,key=getKey,reverse=True)     # Jets are sorted based on their Pt

        InputList=np.zeros((nInputs))
        k=0
        for iList in range(len(Jets_P4)):
            for jList in range(4):
                InputList[k]=Jets_P4[iList][jList]
                k+=1
                if(iList== len(Jets_P4)-1):
                    InputList[k]=MET[0]
                    InputList[k+1]=MET[1]    
    
        if(len(InputList) != 82):
            print "Error -> InputList size:", len(InputList)
    
        minMass=999999.0    
        MJ_Px = None
        MJ1_Px=0
        MJ1_Py=0
        MJ1_Pz=0
        MJ1_E =0
        MJ2_Px=0
        MJ2_Py=0
        MJ2_Pz=0
        MJ2_E =0
   
        for jeti, jet  in enumerate(Jets_P4):
            if not MJ_Px: 
                MJ_Px=jet[1]
                MJ_Py=jet[2]
                MJ_Pz=jet[3]
                MJ_E =jet[0]
            else:
                MJ_Px+=jet[1]
                MJ_Py+=jet[2]
                MJ_Pz+=jet[3]
                MJ_E +=jet[0]   
    
        for nmega1 in range(1,len(Jets_P4)/2+1):
                
            for combination in itertools.combinations(Jets_P4, nmega1):
                
                MJ2_Px=MJ_Px
                MJ2_Py=MJ_Py
                MJ2_Pz=MJ_Pz
                MJ2_E =MJ_E
            
                MJ1_Px=None
                MJ1_Py=None
                MJ1_Pz=None
                MJ1_E =None
            
                for jet in combination:
                
                    MJ2_Px-=jet[1]
                    MJ2_Py-=jet[2]
                    MJ2_Pz-=jet[3]
                    MJ2_E -=jet[0] 
                
                    if MJ1_Px: 
                    
                        MJ1_Px+=jet[1]
                        MJ1_Py+=jet[2]
                        MJ1_Pz+=jet[3]
                        MJ1_E +=jet[0]
                    
                    else: 
                    
                        MJ1_Px=jet[1]
                        MJ1_Py=jet[2]
                        MJ1_Pz=jet[3]
                        MJ1_E =jet[0]
            
                P_MJ1=sqrt((MJ1_Px)**2+(MJ1_Py)**2+(MJ1_Pz)**2)            
                P_MJ2=sqrt((MJ2_Px)**2+(MJ2_Py)**2+(MJ2_Pz)**2)
            
            
                Mass2_MJ1= MJ1_E*MJ1_E - P_MJ1*P_MJ1 
                Mass2_MJ2= MJ2_E*MJ2_E - P_MJ2*P_MJ2
            
                SumMass=Mass2_MJ1+Mass2_MJ2
                if(SumMass < minMass):
                    minMass=SumMass
                    minMJ1 = (MJ1_E, MJ1_Px, MJ1_Py, MJ1_Pz)
                    minMJ2 = (MJ2_E, MJ2_Px, MJ2_Py, MJ2_Pz)
            
    
        # -------------------------------------------------------------------------------- 
        def RazorVariables(MegaJet1, MegaJet2, MET):
            magPj1= sqrt(MegaJet1[1]**2+ MegaJet1[2]**2+MegaJet1[3]**2)
            magPj2= sqrt(MegaJet2[1]**2+ MegaJet2[2]**2+MegaJet2[3]**2)
            Pt_j1 = sqrt(MegaJet1[1]**2 + MegaJet1[2]**2)
            Pt_j2 = sqrt(MegaJet2[1]**2 + MegaJet2[2]**2)
            EtPj12_dot = MET[0] * MegaJet1[1] + MET[1]*MegaJet1[2] + MET[0]*MegaJet2[1] + MET[1]*MegaJet2[2]
            MissingEt = sqrt(MET[0]**2 + MET[1]**2)
       
            M_R = sqrt( (magPj1 + magPj2)**2 - (MegaJet1[3]+MegaJet2[3])**2)   
            MR_T = sqrt( 1.0/2.0 * ( MissingEt * (Pt_j1+Pt_j2) - EtPj12_dot))
            R = MR_T / M_R 
            return M_R,MR_T,R
        # --------------------------------------------------------------------------------
    
        M_R,MR_T,R = RazorVariables (minMJ1,minMJ2,MET)
    
    
        if(M_R*R*R > 0.6):
            nPass1+=1
            target_matrix[iEvent]=1
        else:
            nFail1+=1
            target_matrix[iEvent]=0
    
        
        input_matrix[iEvent]=InputList
        
        R2_Values.append(R*R)
        MR_Values.append(M_R)
        R2_fit.append(0.6/M_R)    
    
print "Pass1=",nPass1,"(" ,nPass1*100.0/(nPass1+nFail1),"%)"
print "Fail1=",nFail1, "(",nFail1*100.0/(nPass1+nFail1),"%)"
print "All =" ,nPass1+nFail1

Pass=[]
for i in range(nPass1):
    Pass.append(1.0)
Fail=[]
for i in range(nFail1):
    Fail.append(2.0)

plt.plot(MR_Values,R2_Values, linestyle='', marker='o')
plt.plot(MR_Values,R2_fit,linestyle='',marker='o',color='r')
plt.xlabel('$M_{R}$')
plt.ylabel('$R^{2}$')
plt.title('$R^{2}$=$f(M_{R})$')
```
