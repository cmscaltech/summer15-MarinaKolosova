```python
from brian2 import *
import brian2.numpy_ as np
import brian2.only as br2
from brian.library.synapses import *
```

```python
import theanets
```

```python
import h5py
import time
import cPickle
%matplotlib inline
```

```python
f=h5py.File("data/data.hdf5")
f.keys()
f['data'].keys()

global input_matrix
global target_matrix
global decorator_matrix

input_matrix     = f['data']['input'].value
target_matrix    = f['data']['target'].value
decorator_matrix = f['data']['deco'].value
f.close()
```

```python
nInputs  = input_matrix.shape[1]
nOutputs = target_matrix.shape[1]
print "Samples in input_matrix=", len(input_matrix), "       Inputs=", nInputs, "       Outputs=",nOutputs
```

```python
truncate=800000
nHiddenNeurons=800
iLayers=4

IN_matrix     = input_matrix[0:truncate,...]      
TARGET_matrix = target_matrix[0:truncate,...]
DEC_matrix    = decorator_matrix[0:truncate,...] 
    
cut   =int(IN_matrix.shape[0]*0.8)
train =IN_matrix[:cut], TARGET_matrix[:cut]
valid =IN_matrix[cut:], TARGET_matrix[cut:]


exp=theanets.feedforward.Regressor(layers=(nInputs,)+tuple([nHiddenNeurons for l in range(iLayers)])+ ( (nOutputs),)) 

fom_train   = 1.0
fom_valid   = 1.0
    
LearningRate=0.0001
Epochs= 0

timing_start = time.mktime(time.gmtime())

for TrainError, ValError in exp.itertrain(train,valid, learning_rate= LearningRate, algo='rprop', 
                                          hidden_dropout=0.9, iteration_size = 400, batch_size=1000, 
                                          validate_every = 1, patience=300, min_improvement = 0.0000001,
                                          rms_halflife=0.99, rms_regularizer=1e-09):
    print 'Epoch=', Epochs+1
    if Epochs%1 == 0 :
        print 'Training   Error = \t',TrainError['err']
        print 'Validation Error = \t',ValError['err']
    Epochs+=1
    if Epochs == 20 :
        break   
            
timing_end=time.mktime(time.gmtime())
#print exp.train(train, valid, learning_rate= LearningRate, algo='rprop', rms_halflife=0.995, rms_regularizer=1e-09)
  
prediction = exp.predict(train[0])
goodness = np.asarray([[cut,(list(train[1][np.where(prediction>cut)]==1).count(True)+list(train[1][np.where(prediction<cut)]==0).count(True))/ float(train[0].shape[0])]  for cut in np.arange(0,1,0.05)])   
fom_train = max( goodness[:,1])    
    
prediction = exp.predict(valid[0])
goodness   = np.asarray([[cut,(list(valid[1][np.where(prediction>cut)]==1).count(True)+list(valid[1][np.where(prediction<cut)]==0).count(True))/ float(valid[0].shape[0])]  for cut in np.arange(0,1,0.05)])
fom_valid  = max( goodness[:,1])     

MaxCut= goodness[np.where(goodness[:,1]==goodness[:,1].max())][0][0]

timing=timing_end-timing_start
    
exp.save('network-pickle.pkl.gz')

print 'Samples=', truncate,'    Hid. Layers=',iLayers, "   Hid. Neurons=", nHiddenNeurons, "   F.O.M. train set=",fom_train, "   FOM valid. set=", fom_valid, " Time=", timing, " Max Cut=", MaxCut      
```

```python
NNtrBit_Agree_Pass=[]      # Trigger Bit = 1 & NN output=1     <-- Keep the information, 4-momenta, R^2, M_R etc.  
NNtrBit_Agree_Fail=[]      # Trigger Bit = 0 & NN output=0
NNtrBit_Disagree_Pass=[]   # Trigger Bit = 1 & NN output=0
NNtrBit_Disagree_Fail=[]   # Trigger Bit = 0 & NN output=1 
```

```python
# prediction -> NN output for training data
# train[0]   -> Training Data
# train[1]   -> Target 

Events_NN1=np.where(prediction>MaxCut)[0]    # Events (indices) with NN output =1 
Events_NN0=np.where(prediction<MaxCut)[0]    # Events (indices) with NN output =0
Events_TB1=np.where(train[1]==1)[0]          # Events (indices) with Trigger Bit =1
Events_TB0=np.where(train[1]==0)[0]          # Events (indices) with Trigger Bit =0

# Search for an element 
def Search(element, SearchArray):
    Found=False
    for i in range(len(SearchArray)):
        if(element==SearchArray[i]):
            Found=True
            return Found
    return Found
```

```python
for i in range(len(Events_TB1)):
    iEvent=Events_TB1[i]
    if(Search(iEvent, Events_NN1)):
        #  Agree NN+TB =1
        NNtrBit_Agree_Pass.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))    
    else:
        #  Disagree TB=1 and NN=0
        NNtrBit_Disagree_Pass.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))

        
for i in range(len(Events_TB0)):
    iEvent=Events_TB0[i]
    
    if(Search(iEvent, Events_NN0)):
        #  Agree NN+TB=0
        NNtrBit_Agree_Fail.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))
    else:
        #  Disagree TB=0 and NN=1
        NNtrBit_Disagree_Fail.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))
```

```python
# Zeros in TB=0, NN=0

print len(NNtrBit_Agree_Fail)
nZeros=1
for i,iAF in enumerate(NNtrBit_Agree_Fail):
    if(sqrt(iAF[0][-1]**2+iAF[0][-2]**2) == 0.0):
        print nZeros ,i, iAF
        nZeros+=1
```

```python
R2_AP =[]
MR_AP =[]
MET_AP=[]
pT_AP =[]

for i, iAP in enumerate(NNtrBit_Agree_Pass):
    R2_AP.append(iAP[1])
    MR_AP.append(iAP[2])
    MET_AP.append(sqrt(iAP[0][-1]**2+iAP[0][-2]**2))
    pT_AP.append(sqrt(iAP[0][1]**2+iAP[0][2]**2))

R2_DP =[]
MR_DP =[]
MET_DP=[]
pT_DP =[]

for i, iDP in enumerate(NNtrBit_Disagree_Pass):
    R2_DP.append(iDP[1])
    MR_DP.append(iDP[2])
    MET_DP.append(sqrt(iDP[0][-1]**2+iDP[0][-2]**2))
    pT_DP.append(sqrt(iDP[0][1]**2+iDP[0][2]**2))
    
R2_AF=[]
MR_AF=[]
MET_AF=[]
pT_AF =[]
for i, iAF in enumerate(NNtrBit_Agree_Fail):
    R2_AF.append(iAF[1])
    MR_AF.append(iAF[2])
    MET_AF.append(sqrt(iAF[0][-1]**2+iAF[0][-2]**2))
    pT_AF.append(sqrt(iAF[0][1]**2+iAF[0][2]**2))
    
R2_DF=[]
MR_DF=[]
MET_DF=[]
pT_DF =[]
for i, iDF in enumerate(NNtrBit_Disagree_Fail):
    R2_DF.append(iDF[1])
    MR_DF.append(iDF[2])
    MET_DF.append(sqrt(iDF[0][-1]**2+iDF[0][-2]**2))
    pT_DF.append(sqrt(iDF[0][1]**2+iDF[0][2]**2))
```
