```python
import h5py
import theanets
import time
import numpy as np
```

```python
f=h5py.File("data/data.hdf5")
f.keys()
f['data'].keys()
input_matrix     = f['data']['input'].value
target_matrix    = f['data']['target'].value
decorator_matrix = f['data']['deco'].value
f.close()
```

```python
mapAlgo=[]
print mapAlgo
```

```python
truncate       =100000

nInputs = input_matrix.shape[1]
nOutputs = target_matrix.shape[1]

input_matrix = input_matrix[0:truncate,...]
target_matrix = target_matrix[0:truncate,...]
decorator_matrix = decorator_matrix[0:truncate,...]

cut   =int(input_matrix.shape[0]*0.8)
train =input_matrix[:cut], target_matrix[:cut]
valid =input_matrix[cut:], target_matrix[cut:]

nHiddenLayers  =1
nHiddenNeurons =400
epochs         =10
trainings      =2

exp=theanets.Experiment(theanets.feedforward.Regressor(layers=(nInputs,)+tuple([nHiddenNeurons for l in range(nHiddenLayers)])+ ( (nOutputs, 'sigmoid'),)),'layerwise')            


timing_start = time.mktime(time.gmtime())
fom = 1.0

for i in range(0, epochs):
    for j in range(trainings):
        print 'Epoch=', i, 'Training=', j, 'Error=', exp.train(train, valid, algo='rmsprop', learning_rate=0.1/10**i) #, momentum=0.95, rms_halflife=0.9 ,rms_regularizer=1e-8) 

timing_end=time.mktime(time.gmtime())

prediction = exp.network.predict(train[0])
goodness = np.asarray([[cut,(list(target_matrix[np.where(prediction>cut)]==1).count(True)+list(target_matrix[np.where(prediction<cut)]==0).count(True))/ float(input_matrix.shape[0])]  for cut in np.arange(0,1,0.05)])
fom = max( goodness[:,1])

    
timing=timing_end-timing_start
print fom, timing
```

```python
MaxCut= goodness[np.where(goodness[:,1]==goodness[:,1].max())][0][0]
print MaxCut
```

```python
NNtrBit_Agree_Pass=[]      # Trigger Bit = 1 & NN output=1     <-- Keep the information, 4-momenta, R^2, M_R etc.  
NNtrBit_Agree_Fail=[]      # Trigger Bit = 0 & NN output=0
NNtrBit_Disagree_Pass=[]   # Trigger Bit = 1 & NN output=0
NNtrBit_Disagree_Fail=[]   # Trigger Bit = 0 & NN output=1 
```

```python
# prediction -> NN output for training data
# train[0]   -> Training Data
# train[1]   -> Target 

Events_NN1=np.where(prediction>MaxCut)[0]    # Events (indices) with NN output =1 
Events_NN0=np.where(prediction<MaxCut)[0]    # Events (indices) with NN output =0
Events_TB1=np.where(train[1]==1)[0]          # Events (indices) with Trigger Bit =1
Events_TB0=np.where(train[1]==0)[0]          # Events (indices) with Trigger Bit =0

# Search for an element 
def Search(element, SearchArray):
    Found=False
    for i in range(len(SearchArray)):
        if(element==SearchArray[i]):
            Found=True
            return Found
    return Found
```

```python
for i in range(len(Events_TB1)):
    iEvent=Events_TB1[i]
    if(Search(iEvent, Events_NN1)):
        #  Agree NN+TB =1
        NNtrBit_Agree_Pass.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))    
    else:
        #  Disagree TB=1 and NN=0
        NNtrBit_Disagree_Pass.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))

        
for i in range(len(Events_TB0)):
    iEvent=Events_TB0[i]
    
    if(Search(iEvent, Events_NN0)):
        #  Agree NN+TB=0
        NNtrBit_Agree_Fail.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))
    else:
        #  Disagree TB=0 and NN=1
        NNtrBit_Disagree_Fail.append((train[0][iEvent],decorator_matrix[iEvent][0], decorator_matrix[iEvent][1], decorator_matrix[iEvent][2]))

```

```python
# Zeros in TB=0, NN=0

print len(NNtrBit_Agree_Fail)
nZeros=1
for i,iAF in enumerate(NNtrBit_Agree_Fail):
    if(sqrt(iAF[0][-1]**2+iAF[0][-2]**2) == 0.0):
        print nZeros ,i, iAF
        nZeros+=1
```

```python
R2_AP =[]
MR_AP =[]
MET_AP=[]
pT_AP =[]

for i, iAP in enumerate(NNtrBit_Agree_Pass):
    R2_AP.append(iAP[1])
    MR_AP.append(iAP[2])
    MET_AP.append(sqrt(iAP[0][-1]**2+iAP[0][-2]**2))
    pT_AP.append(sqrt(iAP[0][1]**2+iAP[0][2]**2))

R2_DP =[]
MR_DP =[]
MET_DP=[]
pT_DP =[]

for i, iDP in enumerate(NNtrBit_Disagree_Pass):
    R2_DP.append(iDP[1])
    MR_DP.append(iDP[2])
    MET_DP.append(sqrt(iDP[0][-1]**2+iDP[0][-2]**2))
    pT_DP.append(sqrt(iDP[0][1]**2+iDP[0][2]**2))
    
R2_AF=[]
MR_AF=[]
MET_AF=[]
pT_AF =[]
for i, iAF in enumerate(NNtrBit_Agree_Fail):
    R2_AF.append(iAF[1])
    MR_AF.append(iAF[2])
    MET_AF.append(sqrt(iAF[0][-1]**2+iAF[0][-2]**2))
    pT_AF.append(sqrt(iAF[0][1]**2+iAF[0][2]**2))
    
R2_DF=[]
MR_DF=[]
MET_DF=[]
pT_DF =[]
for i, iDF in enumerate(NNtrBit_Disagree_Fail):
    R2_DF.append(iDF[1])
    MR_DF.append(iDF[2])
    MET_DF.append(sqrt(iDF[0][-1]**2+iDF[0][-2]**2))
    pT_DF.append(sqrt(iDF[0][1]**2+iDF[0][2]**2))
```
